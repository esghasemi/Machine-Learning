### **Correlation Matrix**

A **correlation matrix** is a table that shows the correlation coefficients between a set of variables. In other words, it shows how strongly two or more variables are related to each other. Correlation coefficients are numerical values that range between -1 and +1, with -1 indicating a perfect negative correlation (i.e., as one variable increases, the other decreases), +1 indicating a perfect positive correlation (i.e., as one variable increases, the other increases), and 0 indicating no correlation.

The correlation matrix is often used in data analysis and statistics to study the relationships between variables. It can help identify which variables are strongly related to each other and which are not, and can be used to explore the underlying structure of a dataset. Correlation matrices can be visualized using heatmaps, which use color to represent the strength and direction of the correlation coefficients.

Correlation matrices are particularly useful in fields such as finance, economics, and social sciences, where researchers often study the relationships between multiple variables. They are also used in machine learning and data science to identify which features (i.e., variables) are most predictive of a target variable, and to identify potential issues such as multicollinearity (i.e., high correlation between two or more predictor variables).

**heatmap() function**:

The heatmap() function in Seaborn allows you to create a heatmap of a matrix or a DataFrame. The function provides a range of arguments to customize the appearance and behavior of the heatmap. Here is an overview of some of the most commonly used arguments:

  **data**: This argument specifies the input data for the heatmap. It can be a matrix or a DataFrame.

  **cmap**: This argument specifies the color map to use for the heatmap. You can choose from a range of built-in color maps such as 'coolwarm', 'viridis', 'magma', 'inferno', and more.

  **center**: This argument specifies the value at which to center the color map. By default, the center is set to 0.

  **annot**: This argument specifies whether to show the data values in each cell of the heatmap. If set to True, the values are displayed; if set to False, they are not.

  **fmt**: This argument specifies the format string for the data values in the heatmap. For example, you can use '.2f' to display the values with two decimal places.

  **linewidths and linecolor**: These arguments specify the width and color of the lines separating the cells in the heatmap.

  **cbar**: This argument specifies whether to show a color bar next to the heatmap, which indicates the color scale.

  **xticklabels and yticklabels**: These arguments specify whether to show the tick labels for the x-axis and y-axis, respectively. If set to True, the labels are displayed; if set to False, they are not. You can also provide a list of labels to use for the ticks.

  **square**: This argument specifies whether to make the cells in the heatmap square. If set to True, the cells are square; if set to False, they are rectangular.

These are just some of the most commonly used arguments for the heatmap() function. There are many other arguments available, such as mask to hide specific cells in the heatmap, vmin and vmax to set the minimum and maximum values for the color map, and more. You can find a full list of arguments in the Seaborn documentation.

```python
import pandas as pd
import seaborn as sns            #Python data visualization library based on Matplotlib (for heatmap)
import numpy as np
import matplotlib.pyplot as plt

float_columns = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity',
                   'Wind Speed (km/h)', 'Wind Bearing (degrees)',
                   'Visibility (km)', 'Pressure (millibars)']

data = df[float_columns]

# Calculate correlation matrix
corr_matrix = data.corr()

# Create heatmap using seaborn
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', linewidths=0.5, annot_kws={"size": 8}, fmt='.2f', yticklabels=corr_matrix.columns)

# Adjust font size of annotations
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show Correlation matrix :
plt.title("Correlation Matrix")
plt.xlabel("Features")
plt.ylabel("hours")
plt.show()

# Adjust margins of PDF file
plt.savefig('CoMatrix.pdf', bbox_inches='tight')
```

## **Part b- LS & RLS**

### **LS**

```python
import numpy as np
import matplotlib.pyplot as plt

class LinearRegressionLS:
    def __init__(self):
        self.coefficients = None

    def fit(self, X, y):
        # Add a column of ones to account for the intercept term
        X = np.column_stack((np.ones(len(X)), X))

        # Compute the coefficients using the least squares method
        self.coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    def predict(self, X):
        # Add a column of ones to account for the intercept term
        X = np.column_stack((np.ones(len(X)), X))

        # Predict the target variable
        return X.dot(self.coefficients)
```

**Humidity & Temperature (C)**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Split the dataset into features (X) and target variable (y)
X_temp = df[['Temperature (C)']].values
y_humidity = df['Humidity'].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_temp, y_humidity, test_size=0.2, random_state=42)

# Initialize and fit the linear regression model using least squares
model = LinearRegressionLS()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred_ls = model.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred_ls)
print("Mean Squared Error:", mse)
```

### **RLS**

```python
import numpy as np
import matplotlib.pyplot as plt

class RecursiveLeastSquares:
    def __init__(self, n_features, forgetting_factor=0.90):
        self.n_features = n_features
        self.forgetting_factor = forgetting_factor
        self.theta = np.zeros((n_features, 1))       # Initialize model parameters
        self.P = np.eye(n_features)                  # Initialize covariance matrix

    def fit(self, X, y):
        errors = []
        for i in range(len(X)):
            x_i = X[i].reshape(-1, 1)
            y_i = y[i]

            # Predict
            y_pred = np.dot(x_i.T, self.theta)

            # Update
            error = y_i - y_pred
            errors.append(error)
            K = np.dot(self.P, x_i) / (self.forgetting_factor + np.dot(np.dot(x_i.T, self.P), x_i))
            self.theta = self.theta + np.dot(K, error)
            self.P = (1 / self.forgetting_factor) * (self.P - np.dot(K, np.dot(x_i.T, self.P)))

        return errors

    def predict(self, X):
        return np.dot(X, self.theta)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Split the dataset into features (X) and target variable (y)
X_temp = df[['Temperature (C)']].values
X_app_temp = df['Apparent Temperature (C)'].values
y_humidity = df['Humidity'].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_temp, y_humidity, test_size=0.2, random_state=42)

# Initialize and fit the RLS model
rls = RecursiveLeastSquares(n_features=X_train.shape[1])
errors = rls.fit(X_train, y_train)

# Make predictions
y_pred_rls = rls.predict(X_test)

# Calculate Mean Squared Error
mse = np.mean(np.array(errors)**2)
print("Mean Squared Error:", mse)

# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_rls, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values for Temperature (C) and Humidity')
plt.grid(True)
plt.show()
```

## **Part c - WLS**

**Temperature & Humidity**

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Split the data for Temperature and Humidity
X_temp = df[['Temperature (C)']].values
X_temp_wls = sm.add_constant(X_temp)

y_humidity = df['Humidity'].values
y_humidity_wls = y_humidity

X_train, X_test, y_train, y_test = train_test_split(X_temp_wls, y_humidity_wls, test_size=0.2, random_state=14)

weights_temp_train = np.ones_like(y_train)

wls_model_temp_train = sm.WLS(y_train, X_train, weights=weights_temp_train).fit()

y_pred_wls_temp = wls_model_temp_train.predict(X_test)

mse_wls_temp = mean_squared_error(y_test, y_pred_wls_temp)
print(f'Mean Squared Error of WLS: {mse_wls_temp}')

plt.scatter(X_test[:, 1], y_test, color='blue', alpha=0.5, label='Actual Temperature')
plt.plot(X_test[:, 1], y_pred_wls_temp, color='red', label='WLS Predicted Humidity', linewidth=2)
plt.title('WLS Estimation of Temperature vs. Humidity (Test Set)')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity')
plt.legend()
plt.show()
```
