## **Data Spliting**

```python
#x=df.iloc[:,:-1].values
#label=df.iloc[:,-1].values
x = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=42)
print("x_train shape",x_train.shape)
print("x_test shape",x_test.shape)
print("y_train shape",y_train.shape)
print("y_test shape",y_test.shape)
```

## **Gaussian Naive Bayes**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, confusion_matrix, f1_score , classification_report

import seaborn as sns


# Initialize Gaussian Naive Bayes
model = GaussianNB()

# Train the model
model.fit(x_train, y_train)

# Predict the test set results
y_pred = model.predict(x_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='binary')

print("Classification Report (Linear Regression):\n", classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)

# Plotting the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis', cbar_kws={'shrink': .75})
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

## **5 Random Data**

```python
# Converting Numpy array to Pandas Dataframe
x_test_df = pd.DataFrame(x_test, columns=x.columns).reset_index(drop=True)
y_test_df = pd.Series(y_test).reset_index(drop=True)

# 5 Random data from Test
random_indices = np.random.choice(x_test_df.index, size=5, replace=False)
random_samples = x_test_df.loc[random_indices]
random_predictions = model.predict(random_samples)
random_actuals = y_test_df.loc[random_indices]

# Cpmoarision
for i in range(5):
    print(f"Sample {i+1}:")
    print(f"Features: {random_samples.iloc[i].tolist()}")
    print(f"Actual Value: {random_actuals.iloc[i]}")
    print(f"Predicted Value: {random_predictions[i]}")
    print()
```

```python
random_indices
```

```python
comparison_df = pd.DataFrame({
    'Index': random_indices,
    'Actual': random_actuals.values,
    'Predicted': random_predictions
})

plt.figure(figsize=(12, 6))
comparison_df.set_index('Index')[['Actual', 'Predicted']].plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Comparison of Actual and Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.xticks(rotation=0)
plt.legend(['Actual', 'Predicted'])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(comparison_df['Index'], comparison_df['Actual'], color='b', label='Actual', s=100)
plt.scatter(comparison_df['Index'], comparison_df['Predicted'], color='r', label='Predicted', s=100, marker='x')
plt.title('Comparison of Actual and Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
